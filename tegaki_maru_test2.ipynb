```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

tar = 'datasets/tegaki/maru/maru_1_00057.png'
im = cv2.imread(tar)
cascade = cv2.CascadeClassifier('TrainingAssistant/results/cascades/tegaki_maru2/cascade.xml') #
circles= cascade.detectMultiScale(im, 1.1, 3) #

for (x, y, w, h) in circles:
    upper_left = (x, y)
    bottom_right = (x+w, y+h)
    cv2.rectangle(im, upper_left, bottom_right, (255, 20, 147), thickness=3) #

# BGRRGB
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

# BGRRGB
plt.imshow(im)
plt.show()

# cv2.imwrite('result.jpg', im)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 
lower_yellow = np.array([100,100,100])
upper_yellow = np.array([150,150,150])

cap = cv2.VideoCapture(0)
 
cascade = cv2.CascadeClassifier('TrainingAssistant/results/cascades/tegaki_maru2/cascade.xml') #

while(1):
    # 
    ret, frame = cap.read()
 
    # HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
 
    circles= cascade.detectMultiScale(frame, 1.1, 3) #

    for (x, y, w, h) in circles:
        upper_left = (x, y)
        bottom_right = (x+w, y+h)
        cv2.rectangle(frame, upper_left, bottom_right, (255, 20, 147), thickness=3) #
 
    cv2.imshow("display", frame)
 
    # q
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
 
cap.release()
cv2.destroyAllWindows()

```
