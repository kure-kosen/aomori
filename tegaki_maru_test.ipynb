```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

tar = 'datasets/tegaki/maru/maru00500.png'
im = cv2.imread(tar)
cascade = cv2.CascadeClassifier('TrainingAssistant/tegaki_maru/cascade.xml') #
circles= cascade.detectMultiScale(im, 1.1, 3) #

for (x, y, w, h) in circles:
    upper_left = (x, y)
    bottom_right = (x+w, y+h)
    cv2.rectangle(im, upper_left, bottom_right, (255, 20, 147), thickness=3) #

# BGRRGB
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

# BGRRGB
plt.imshow(im)
plt.show()

# cv2.imwrite('result.jpg', im)
```
